```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# read file data
data = pd.read_csv('data_linear.csv').values
# in so hang trong array numpy matrix shape(m hang,n cot) shape[0] => so hang
N = data.shape[0]

#tao matran 1*n dien tich
x = data[:,0].reshape(-1,1)

#tao matran 1*n gia tien tuong ung
y = data[:,1].reshape(-1,1)

plt.scatter(x, y)
plt.xlabel('mét vuông')
plt.ylabel('giá')

# tao mot ma tran cac gia tri 1 su dung ham n.ones va su dung ham hstack de ghep 2 matran voi nhau
x = np.hstack((np.ones((N, 1)), x))
print(x)

# tao duong thang gia dinh trung voi truc Ox y = x
w = np.array([0.,1.]).reshape(-1,1)
print(w)

# Thuat toan Gradient descent
numOfIteration = 100

#tao ma tran cost 30 dong voi tat ca gia tri = 0
cost = np.zeros((numOfIteration,1))
learning_rate = 0.000001

# su dung gradient descent tim MIN loss function : cost
for i in range(1, numOfIteration):
    #tính loss (y'-y)
    r = np.dot(x, w) - y
    #ham f(x)
    cost[i] = 0.5*np.sum(r*r)
    #Tim cuc tri dua vao buoc 2 cua thuat toan gradient
    w[0] -= learning_rate*np.sum(r)
    # correct the shape dimension
    w[1] -= learning_rate*np.sum(np.multiply(r, x[:,1].reshape(-1,1)))
    print(cost[i])
# print(r)
predict = np.dot(x, w)

plt.plot((x[0][1], x[N-1][1]),(predict[0], predict[N-1]), 'r')
plt.show()
x1 = 50
y1 = w[0] + w[1] * x1

print(w)
print('Giá nhà cho 50m^2 là : ', y1)
```
